```python
reviews_data["rating"] = reviews_data["rating"].replace({'A+':'15/15',
                                                 'A':'14/15',
                                                 'A-':'13/15',
                                                 'B+':'12/15',
                                                 'B':'11/15',
                                                 'B-':'10/15',
                                                 'C+':'9/15',
                                                 'C':'8/15',
                                                 'C-':'7/15',
                                                 'D+':'6/15',
                                                 'D':'5/15',
                                                 'D-':'4/15',
                                                 'F+':'3/15',
                                                 'F':'2/15',
                                                 'F-':'1/15'})
```

```python
import string 
letters = string.ascii_uppercase

def to_decimal(fraction):
    if isinstance(fraction, str):
        try:
            a, b = fraction.split('/')
            x  = round((float(a) / float(b)), 2)
            return x
        except ValueError:
            return fraction
    else:
        return fraction
    
reviews_data['rating'] = reviews_data['rating'].map(lambda x: to_decimal(x) if x is not letters else x )
reviews_data['rating']
```
