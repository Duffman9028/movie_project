We already found evidence that movies in the drama, thriller, and horror genres can yield very high ROIs.  This is because among movies with the highest ROIs (over 50), those genres appear significantly more than the other genres.  This is important because, as a new studio, it is important to get a high return on investment.  In the short term, the studio should focus on dramas, thrillers, and horrors in order to limit budget, maximize gross profit, and get the most bang for its buck.  

In the spirit of maximizing ROI in the early days of the studio, we want to hire directors and writers that are particularly trendy right now in order to generate buzz for our movies.  

```python
import pandas as pd
```

```python
#pulling directors dataframe from IMDB 
dir_df = pd.read_csv('directors.csv')
```

```python
dir_df.info()
```

```python
#Since we're only interested in drama, horror, and thrillers, we will filter for movies in at least one of those genres.
dir_df2 = dir_df[dir_df.genres.str.contains('Drama|Thriller|Horror', na=False)].reset_index(drop=True)
dir_df2.head(20)
```

```python
#We will grab popularity data from the TMDB dataset
tmdb_df = pd.read_csv("zippedData/tmdb.movies.csv.gz", compression = 'gzip', index_col=0)
tmdb_df.head()
```

```python
tmdb_df.info()
```

```python
tmdb_df = tmdb_df.drop_duplicates(subset=['original_title'])
```

```python
tmdb_df.info()
```

```python
director_df = dir_df2.merge(tmdb_df, left_on='primary_title', right_on='title',how='inner')[['director_name','primary_title','genres', 'popularity']]
```

```python
dir_pop_df = director_df.groupby(['director_name']).sum().sort_values(by='popularity', ascending=False)
```

```python
dir_pop_df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(data = dir_pop_df.head(20), y = dir_pop_df.head(20).index, x = 'popularity',). \
set(title = "Trending Directors (Drama, Horror, & Thrillers)", xlabel= 'Popularity', ylabel = 'Director Name')
plt.tight_layout()
fig.savefig('director.png');
```

Denis Villeneuve, Chad Stahelski, Ridley Scott, and David Leitch are the most trendy directors of movies from these genres.

```python
#Want to see what movies the above directors directed.
director_df[director_df.director_name.isin(["Denis Villeneuve", "Chad Stahelski", "Ridley Scott", "David Leitch"])]
```

Some trending movies for each director: <br>
Ridley Scott:  Exodus: Gods and Kings, Alien: Covenant <br>
David Leitch:  John Wick <br>
Chad Stahelski:  John Wick, John Wick Ch. 2 <br>
Denis Villeneuve: Blade Runner 2049 <br>


### Writers

Repeating this process for writers:

```python
#pulling directors dataframe from IMDB 
wri_df = pd.read_csv('writers.csv')
```

```python
wri_df.info()
```

```python
#Since we're only interested in drama, horror, and thrillers, we will filter for movies in at least one of those genres.
wri_df2 = wri_df[wri_df.genres.str.contains('Drama|Thriller|Horror', na=False)].reset_index(drop=True)
wri_df2.head(20)
```

```python
tmdb_df.info()
```

```python
writer_df = wri_df2.merge(tmdb_df, left_on='primary_title', right_on='title',how='inner')[
    ['writer_name','primary_title','genres', 'popularity']]
```

```python
wri_pop_df = writer_df.groupby(['writer_name']).sum().sort_values(by='popularity', ascending=False)
```

```python
wri_pop_df.head()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(data = wri_pop_df.head(20), y = wri_pop_df.head(20).index, x = 'popularity',). \
set(title = "Trending Writers (Drama, Horror, & Thrillers)", xlabel= 'Popularity', ylabel = 'Writer Name')
plt.tight_layout()
fig.savefig('writer.png');
```

Derek Kolstad, Eric Heisserer, Sylvester Stallone, and Hossein Amini are the most trendy directors of movies from these genres.

```python
#Want to see what movies the above writers wrote.
writer_df[writer_df.writer_name.isin(["Derek Kolstad", "Eric Heisserer", "Sylvester Stallone", "Hossein Amini"])].sort_values(
    'writer_name')
```

Some trending movies for each writer: <br>
Derek Kolstad:  John Wick, John Wick Ch 2 <br>
Eric Heisserer:  Bird Box, Arrival <br>
Hossein Amini:  47 Ronin, Snow White and the Huntsman <br>
Sylvester Stallone: Creed II, The Expendables <br>
