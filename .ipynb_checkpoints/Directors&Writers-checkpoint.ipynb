# Trending Directors and Writers

We already found evidence that movies in the horror, mystery, thriller genres can yield very high ROIs.  This is because among movies with the highest ROIs (over 50), those genres appear significantly more than the other genres.  This is important because, as a new studio, it is important to get a high return on investment.  In the short term, the studio should focus on dramas, thrillers, and horrors in order to limit budget, maximize gross profit, and get the most bang for its buck.  

In the spirit of maximizing ROI in the early days of the studio, we want to hire directors and writers that are particularly trendy right now in order to generate buzz for our movies.  

```python
import pandas as pd
```

### Directors

```python
#pulling directors dataframe from IMDB 
dir_df = pd.read_csv('directors.csv')
```

```python
dir_df.info()
```

```python
#Since we're only interested in drama, horror, and thrillers, we will filter for movies in at least one of those genres.
dir_df2 = dir_df[dir_df.genres.str.contains('Mystery|Thriller|Horror', na=False)].reset_index(drop=True)
dir_df2.head(20)
```

```python
#We will grab popularity data from the TMDB dataset
tmdb_df = pd.read_csv("zippedData/tmdb.movies.csv.gz", compression = 'gzip', index_col=0)
tmdb_df.head()
```

```python
tmdb_df.info()
```

```python
#There are some duplicate rows in the TMDB data set, so we will drop duplicates so there is only 1 row per movie
tmdb_df = tmdb_df.drop_duplicates(subset=['original_title'])
```

```python
tmdb_df.info()
```

```python
#Here we will create a DataFrame with director names, movie titles, genres, and popularity.
#Popularity is a statistic from the TMDB dataset that measures the popularity of a movie  based on, among other things,
#webpage views, and changes daily.
#We will do this by merging the tmdb DataFrame we just pulled with the filtered IMDB DataFrame created above, on movie title
director_df = dir_df2.merge(tmdb_df, left_on='primary_title', right_on='title',how='inner')[[
    'director_name','primary_title','genres', 'popularity']]
director_df.head()
```

```python
#Now we group by director name and sum the popularity column.  This creates a popularity sum statistic that is equal to the 
#sum of all the movies in the selected genres for each director.  This is an indication of how much buzz a director has 
#for all the movies from these genres.  
dir_pop_df = director_df.groupby(['director_name']).sum().sort_values(by='popularity', ascending=False).rename(columns= \
                {'popularity':'popularity_sum'})
                                                                                                              
```

```python
dir_pop_df.head()
```

#### Creating the viz

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
#Creating a horizontal bar graph showing popularity sum for each director
fig, ax = plt.subplots(figsize=(8,6))

#Using only the top 20 directors
sns.barplot(data = dir_pop_df.head(20), y = dir_pop_df.head(20).index, x = 'popularity_sum',). \
set(xlabel= 'Popularity Sum', ylabel = 'Director Name')

#Creating Title and subtitle 
ax.text(x=0.5, y=1.1, s='Trending Directors', fontsize=20, weight='bold', ha='center', va='bottom', transform=ax.transAxes)
ax.text(x=0.5, y=1.05, s='Horror, Thriller, & Mystery', fontsize=8, alpha=0.75, ha='center', va='bottom', 
        transform=ax.transAxes)
plt.tight_layout()

#Saving the viz to the repo
fig.savefig('director.png');
```

Denis Villeneuve, Chad Stahelski, James Wan, and David Leitch are the most trendy directors of movies from these genres.

```python
#Want to see what movies the above directors directed.
director_df[director_df.director_name.isin(["Denis Villeneuve", "Chad Stahelski", "James Wan", "David Leitch"])]
```

Some trending movies for each director: <br>
James Wan:  Furious 7 <br>
David Leitch:  John Wick <br>
Chad Stahelski:  John Wick, John Wick Ch. 2 <br>
Denis Villeneuve: Blade Runner 2049 <br>


### Writers

Repeating this process for writers:

```python
#pulling directors dataframe from IMDB 
wri_df = pd.read_csv('writers.csv')
```

```python
wri_df.info()
```

```python
#Since we're only interested in mystery, horror, and thrillers, we will filter for movies in at least one of those genres.
wri_df2 = wri_df[wri_df.genres.str.contains('Mystery|Thriller|Horror', na=False)].reset_index(drop=True)
wri_df2.head(20)
```

```python
tmdb_df.info()
```

```python
#Here we will create a DataFrame with writer names, movie titles, genres, and popularity.
#Popularity is a statistic from the TMDB dataset that measures the popularity of a movie  based on, among other things,
#webpage views, and changes daily.
#We will do this by merging the tmdb DataFrame with the filtered IMDB DataFrame created above, on movie title
writer_df = wri_df2.merge(tmdb_df, left_on='primary_title', right_on='title',how='inner')[
    ['writer_name','primary_title','genres', 'popularity']]
```

```python
#Now we group by writer name and sum the popularity column.  This creates a popularity sum statistic that is equal to the 
#sum of all the movies in the selected genres for each writer.  This is an indication of how much buzz a writer has 
#for all the movies from these genres.  
wri_pop_df = writer_df.groupby(['writer_name']).sum().sort_values(by='popularity', ascending=False).rename(columns= \
                {'popularity':'popularity_sum'})
```

```python
wri_pop_df.head()
```

```python
#Creating a horizontal bar graph showing popularity sum for each director
fig, ax = plt.subplots(figsize=(8,6))

#Using only the top 20 writers
sns.barplot(data = wri_pop_df.head(20), y = wri_pop_df.head(20).index, x = 'popularity_sum',). \
set(xlabel= 'Popularity Sum', ylabel = 'Writer Name')

#Creating Title and subtitle 
ax.text(x=0.5, y=1.1, s='Trending Writers', fontsize=20, weight='bold', ha='center', va='bottom', transform=ax.transAxes)
ax.text(x=0.5, y=1.05, s='Horror, Thriller, & Mystery', fontsize=8, alpha=0.75, ha='center', va='bottom', 
        transform=ax.transAxes)
plt.tight_layout()

#Saving the viz to the repo
fig.savefig('writer.png');
```

Derek Kolstad, Eric Heisserer, Luc Besson, and Richard Wenk are the most trendy directors of movies from these genres.

```python
#Want to see what movies the above writers wrote.
writer_df[writer_df.writer_name.isin(["Derek Kolstad", "Eric Heisserer", "Luc Besson", "Richard Wenk"])].sort_values(
    'writer_name')
```

Some trending movies for each writer: <br>
Derek Kolstad:  John Wick, John Wick Ch 2 <br>
Eric Heisserer:  Bird Box, Arrival <br>
Luc Besson:  Taken 3 <br>
Richard Wenk: The Equalizer, The Equalizer 2 <br>
