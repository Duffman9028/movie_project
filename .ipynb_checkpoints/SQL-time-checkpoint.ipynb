```python
import sqlite3
import pandas as pd 

conn = sqlite3.Connection("im.db")
```

Looking into which director would maximize user ratings and ROI.

```python

director_df = pd.read_sql("""
                            SELECT movie_id, mr.averagerating, mr.numvotes, p.primary_name, p.death_year
                            FROM movie_ratings AS mr
                                JOIN directors AS d
                                    USING(movie_id)
                                JOIN persons AS p
                                    USING(person_id)
                            GROUP BY 
    """,conn)
```

```python
director_df.drop_duplicates(inplace=True)
director_df
```

```python
#Keeping only directors who are still alive
director_df = director_df[director_df.death_year.isna()]
```

```python
director_df.groupby(by='primary_name').agg('mean','sum')
```
